# ML Theory examples and visualizations

## convex and non-convex

```python
import numpy as np
import matplotlib.pyplot as plt

# Define a convex function (e.g., a quadratic function)
def convex_function(x):
    return x**2

# Define a non-convex function (e.g., a sine-modulated polynomial)
def non_convex_function(x):
    return np.sin(2 * np.pi * x) + 0.5 * x**2

# Generate x values
x = np.linspace(-2, 2, 400)

# Compute function values
y_convex = convex_function(x)
y_non_convex = non_convex_function(x)

# Plot both functions
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

# Convex function plot
axs[0].plot(x, y_convex, 'b', label=r'$f(x) = x^2$')
axs[0].set_title("Convex Function")
axs[0].legend()
axs[0].grid()

# Non-convex function plot
axs[1].plot(x, y_non_convex, 'r', label=r'$f(x) = \sin(2\pi x) + 0.5x^2$')
axs[1].set_title("Non-Convex Function")
axs[1].legend()
axs[1].grid()

plt.tight_layout()
plt.show()
```

## Gradient descent in convex and non-convex functions

```python
# Define a convex function and its gradient
def convex_function(x):
    return x**2

def convex_gradient(x):
    return 2 * x

# Define a non-convex function and its gradient
def non_convex_function(x):
    return np.sin(2 * np.pi * x) + 0.5 * x**2

def non_convex_gradient(x):
    return 2 * np.pi * np.cos(2 * np.pi * x) + x  # Gradient of sin(2x) + 0.5x^2

# Gradient descent parameters
learning_rate = 0.1
iterations = 20

# Initial points
x_convex = 1.5
x_non_convex = -1.5

# Store values for visualization
convex_path = [x_convex]
non_convex_path = [x_non_convex]

for _ in range(iterations):
    x_convex -= learning_rate * convex_gradient(x_convex)
    x_non_convex -= learning_rate * non_convex_gradient(x_non_convex)
    convex_path.append(x_convex)
    non_convex_path.append(x_non_convex)

# Generate x values for plotting
x_vals = np.linspace(-2, 2, 400)
y_convex_vals = convex_function(x_vals)
y_non_convex_vals = non_convex_function(x_vals)

# Convert paths to numpy arrays for plotting
convex_path = np.array(convex_path)
non_convex_path = np.array(non_convex_path)

# Plot both functions and gradient descent steps
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Convex function plot
axs[0].plot(x_vals, y_convex_vals, 'b', label=r'$f(x) = x^2$')
axs[0].scatter(convex_path, convex_function(convex_path), color='red', marker='o', label='Gradient Descent Steps')
axs[0].set_title("Convex Function Optimization")
axs[0].legend()
axs[0].grid()

# Non-convex function plot
axs[1].plot(x_vals, y_non_convex_vals, 'r', label=r'$f(x) = \sin(2\pi x) + 0.5x^2$')
axs[1].scatter(non_convex_path, non_convex_function(non_convex_path), color='red', marker='o', label='Gradient Descent Steps')
axs[1].set_title("Non-Convex Function Optimization")
axs[1].legend()
axs[1].grid()

plt.tight_layout()
plt.show()
```
